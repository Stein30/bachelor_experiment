```python
import numpy as np

file_name =  'output2.xlsx'
import pandas as pd

dfx= pd.read_excel(io=file_name, usecols=[3,7,8,9,10])
df_ss = dfx[dfx['geo'].str.contains('Ss')] #砂岩

print(len(df_ss))

a = df_ss.to_numpy()

l = len(a)
j = 0
maxj = 0
a[0, 4] = 'Ss' + str(j)

for i in range(1,l):
    if a[i, 0] == a[i-1, 0] and a[i, 1] == a[i-1, 1]:
        j = j + 1
        a[i, 4] = 'Ss' + str(j)
        if maxj < j:
            maxj = j
    else:
        j = 0
        a[i, 4] = 'Ss' + str(j)
        
print(maxj)

column = ['lat', 'lon', 'alt', 'depth', 'geo']
dfs = pd.DataFrame(data=a, columns=column)

for i in range(maxj):
    geo = 'Ss' + str(i)
    dfxs = dfs[dfs['geo'].isin([geo])] #砂岩
    
    dfx_ss = dfxs.drop(['geo','depth'], axis=1)
    dfx_y = dfxs.drop(['lat','lon','alt','geo'], axis=1)
    X3 = dfx_ss.to_numpy()
    y = dfx_y.to_numpy()
    
    print(geo, len(dfxs))
    print(dfxs['depth'].max(), dfxs['depth'].min())
    print(dfxs['depth'].describe())
```
